```python
cats_df = df[df['animal_type'] == 'Cat'].reset_index()
del cats_df['index']
print(len(cats_df))
cats_df.head()
```

```python
cats_df.sort_values(by=['age_upon_outcome'])
```

```python
print(cats_df['sex_upon_outcome'].unique())
```

```python
cats_df['sex'] = np.where(cats_df['sex_upon_outcome'].str.contains('Male'), 'Male', 'Female')
cats_df['is_sterilized'] = np.where(cats_df['sex_upon_outcome'].str.contains('Intact'), 0, 1)
```

## EDA (cats)

```python
import seaborn as sns
```

```python
cats_df['Cat/Kitten (outcome)'] = np.where(cats_df['age_upon_outcome_(days)'] >= 180, 'Cat', 'Kitten')
```

```python
g = sns.factorplot(x='outcome_type', hue='Cat/Kitten (outcome)', col='sex_upon_outcome', 
               col_wrap=3, data=cats_df, kind='count', size=4.5)

g.set_xticklabels(rotation=90)
g.set_xlabels('Outcome')
```
